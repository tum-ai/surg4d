```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
output_path = Path('output/cholecseg8k/video01_00080_qwen_cat/video_lang/ours_10000')
render_path = output_path / 'renders_npy'
gt_path = output_path / 'gt_npy'
```

```python
frame = 50
gt = np.load(gt_path / f'{frame:05}.npy')
render = np.load(render_path / f'{frame:05}.npy')
gt_patch, gt_inst = gt[..., :3], gt[..., 3:]
render_patch, render_inst = render[..., :3], render[..., 3:]
```

```python
col_transform = lambda c: c/2+0.5
```

```python
fig, axs = plt.subplots(2, 2, figsize=(20, 15))
axs[0, 0].imshow(col_transform(gt_patch))
axs[0, 0].set_title("patch ground truth")
axs[0, 1].imshow(col_transform(gt_inst))
axs[0, 1].set_title("instance ground truth")
axs[1, 0].imshow(col_transform(render_patch))
axs[1, 0].set_title("patch render")
axs[1, 1].imshow(col_transform(render_inst))
axs[1, 1].set_title("instance render")
plt.show()
```
