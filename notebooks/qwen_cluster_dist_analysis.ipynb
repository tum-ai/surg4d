```python
import torch
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
from plyfile import PlyData
```

```python
clusters_path = Path('output/hypernerf/chickchicken_qwen/chickchicken_qwen_0_base/graph/clusters.npy')
qwen_features_path = Path('output/hypernerf/chickchicken_qwen/chickchicken_qwen_0_base/graph/latent_qwen_filtered.npy')
```

```python
clusters = np.load(clusters_path)
lf = np.load(qwen_features_path)
```

```python
lf_colors = lf * 10
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(*lf_colors[clusters == 15].T, c=np.clip(lf_colors[clusters == 15], a_min=0, a_max=1))
ax.set_title(f'Cluster {15}')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(*lf_colors[clusters == 15][:100].T, c=np.clip(lf_colors[clusters == 15][:100], a_min=0, a_max=1))
ax.set_title(f'Cluster {15}')
plt.show()
```

```python
for cluster_id in np.unique(clusters):
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.scatter(*lf[clusters == cluster_id].T, c=np.clip(lf_colors[clusters == cluster_id], a_min=0, a_max=1))
    ax.set_title(f'Cluster {cluster_id}')
plt.show()
```
